# Phase 2 start: Model Training and Evaluation

In this phase, we will train a machine learning model using the cleaned dataset from Phase 1. We will also evaluate the model's performance using various metrics.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('data/Algerian_forest_fires_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
##drop month,day and yyear
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## Encoding
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df.tail()
```

```python
df['Classes'].value_counts()
```

```python
## Independent And dependent features
X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
X.head()
```

```python
y
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
## Feature Selection based on correlaltion
X_train.corr()
```

```python
## Check for multicollinearity
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
plt.title('Correlation Heatmap', fontsize=20)
plt.show()
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold--Domain expertise
corr_features=correlation(X_train,0.85)
```

```python
## drop features when correlation is more than 0.85 
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

## Feature Scaling Or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

## Box Plots To understand Effect Of Standard Scaler

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
sns.boxplot(data=X_train, ax=axes[0])
axes[0].set_title('X_train Before Scaling')
sns.boxplot(data=pd.DataFrame(X_train_scaled, columns=X_train.columns), ax=axes[1])
axes[1].set_title('X_train After Scaling')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression, Lasso, LassoCV, Ridge, RidgeCV, ElasticNet, ElasticNetCV
from sklearn.metrics import mean_absolute_error, r2_score
```

## Linear Regression Model

```python

linreg = LinearRegression()
linreg.fit(X_train_scaled, y_train)
y_pred = linreg.predict(X_test_scaled)
print("Mean absolute error", mean_absolute_error(y_test, y_pred))
print("R2 Score", r2_score(y_test, y_pred))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual FWI")
plt.ylabel("Predicted FWI")
plt.title("Linear Regression: Actual vs Predicted FWI")
plt.show()
```

## Lasso Regression

```python
# Lasso Regression with optimized code (no redundant imports)
lasso = Lasso()
lasso.fit(X_train_scaled, y_train)
y_pred = lasso.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y_test, y_pred)
plt.xlabel("Actual FWI")
plt.ylabel("Predicted FWI")
plt.title("Lasso Regression: Actual vs Predicted FWI")
plt.show()
```

### Cross Validation Lasso

```python
lassocv = LassoCV(cv=5)
lassocv.fit(X_train_scaled, y_train)
```

## Ridge Regression model

```python
ridge = Ridge()
ridge.fit(X_train_scaled, y_train)
y_pred = ridge.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y_test, y_pred)
plt.xlabel("Actual FWI")
plt.ylabel("Predicted FWI")
plt.title("Ridge Regression: Actual vs Predicted FWI")
plt.show()
```

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(X_train_scaled,y_train)
y_pred=ridgecv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
plt.title("Ridge RegressionCV: Actual vs Predicted FWI")
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```
